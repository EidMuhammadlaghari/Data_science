```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Bivariate Exploration of Data

## Scatterplots

```python
plt.scatter(data=diamond, x='carat', y='price')
```

```python
plt.scatter(data=diamond, x='carat', y='price')
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
sns.regplot(data=diamond, x='carat', y='price')
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
sns.boxplot(data=diamond, x='cut', y='price')
```

## Heat Maps

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data=diamond.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlation between Numeric Features in Diamond')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist2d(data=diamond, x='carat', y='price', cmin=0.5)
plt.colorbar()
```

```python
import matplotlib.pyplot as plt

plt.hist2d(data=diamond, x='carat', y='price', cmin=0.5,cmap='viridis_r')
plt.colorbar()
```

## Boxplot

```python
base_color = sns.color_palette()[0]

# Create a box plot for 'cut' vs. 'price' in diamonds
sns.boxplot(data=diamond, x='cut', y='price', color=base_color)
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Box Plot of Cut vs. Price in Diamond')
```

## Clustered Bar Charts

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=diamond, x='cut', y='price', hue='clarity', palette='Set1')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Clustered Bar Chart of Cut vs. Price in Diamond by Clarity')
plt.legend(title='Clarity', title_fontsize='12', loc='upper right', bbox_to_anchor=(1.2, 1))
```

## Line plot

```python
plt.figure(figsize=(10, 6))
sns.pointplot(data=diamond, x='cut', y='price', ci='sd', color='b')
plt.xlabel('Cut')
plt.ylabel('Average Price')
```
